```python
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
#LOAD MODEL FOR PREDICTION

model=load_model('model.h5')

with open('encoder.pkl','rb') as file:
    oheencodergeo=pickle.load(file)
    
with open('encodegender.pkl','rb') as file:
    labelencodergender=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
    
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
#encoding Geography

encoded_geo=oheencodergeo.transform([[input_data['Geography']]]).toarray()
encoded_geo_df=pd.DataFrame(encoded_geo,columns=oheencodergeo.get_feature_names_out(['Geography']))
encoded_geo_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
#combine with input data

input_df=pd.concat([input_df.drop("Geography",axis=1),encoded_geo_df],axis=1)
input_df
```

```python
##labeltransform

input_df['Gender']=labelencodergender.transform(input_df['Gender'])
input_df
```

```python
##scalling the data

scaled_input=scaler.transform(input_df)
scaled_input
```

```python
##Prediction

prediction=model.predict(scaled_input)

prediction
```

```python
if(prediction[0][0]>0.5):
    print("Customer will exit.")
    
else:
    print("Customer will stay.")
```
